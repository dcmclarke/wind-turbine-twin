```python
import h5py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
DATA_PATH = Path("C:/Projects/win-turbine-twin/data/aventa_rotor_icing/")

# Find the HDF5 file
hdf5_files = list(DATA_PATH.glob("*.hdf5")) + list(DATA_PATH.glob("*.h5"))
print("Found files:", hdf5_files)
```

```python
# Open the first file and print everything inside it
hdf5_path = hdf5_files[0]
print(f"Opening: {hdf5_path.name}\n")

with h5py.File(hdf5_path, "r") as f:
    def print_structure(name, obj):
        indent = "  " * name.count("/")
        if isinstance(obj, h5py.Dataset):
            print(f"{indent}{name}: shape={obj.shape}, dtype={obj.dtype}")
        else:
            print(f"{indent}{name}/")
    
    f.visititems(print_structure)
```

```python
with h5py.File(hdf5_path, "r") as f:
    # Get the first time block name
    first_block = list(f["Aventa"].keys())[0]
    print(f"First block: {first_block}")
    
    # Read the channel list - these are byte strings
    channel_list = f[f"Aventa/{first_block}/ChannelList"][:]
    print("\n--- Channel names ---")
    for i, ch in enumerate(channel_list):
        print(f"  {i}: {ch[0].decode('utf-8')}")
    
    # Read one WM channel to see what's in it
    print("\n--- WM1 sample values ---")
    wm1_values = f[f"Aventa/{first_block}/WM1/Value"][:]
    wm1_time = f[f"Aventa/{first_block}/WM1/Time"][:]
    print(f"  Value shape: {wm1_values.shape}")
    print(f"  First 10 values: {wm1_values[:10].flatten()}")
    print(f"  Min: {wm1_values.min():.3f}, Max: {wm1_values.max():.3f}")
    
    # Read temperature
    print("\n--- ATM_TEMP_01 sample ---")
    temp = f[f"Aventa/{first_block}/ATM_TEMP_01/Value"][:]
    print(f"  Shape: {temp.shape}")
    print(f"  First 5 values: {temp[:5].flatten()}")
    print(f"  Min: {temp.min():.2f}°C, Max: {temp.max():.2f}°C")
```

```python
with h5py.File(hdf5_path, "r") as f:
    first_block = list(f["Aventa"].keys())[0]
    
    print("--- WM channel ranges (what are these signals?) ---")
    for wm in ["WM1", "WM2", "WM3", "WM4", "WM5"]:
        vals = f[f"Aventa/{first_block}/{wm}/Value"][:].flatten()
        # Filter out obvious sentinel values
        valid = vals[vals > -999]
        print(f"  {wm}: min={valid.min():.2f}, max={valid.max():.2f}, mean={valid.mean():.2f}")

# Now check the December icing file - this is where we expect cold temps
print("\n--- Checking December icing file ---")
dec_file = [f for f in hdf5_files if "17_12" in f.name][0]
print(f"File: {dec_file.name}")

with h5py.File(dec_file, "r") as f:
    blocks = list(f["Aventa"].keys())
    print(f"Blocks in this file: {blocks[:5]}...")
    
    first_block = blocks[0]
    temp = f[f"Aventa/{first_block}/ATM_TEMP_01/Value"][:].flatten()
    valid_temp = temp[temp > -999]
    print(f"\nTemperature in December file:")
    print(f"  Min: {valid_temp.min():.2f}°C")
    print(f"  Max: {valid_temp.max():.2f}°C") 
    print(f"  Mean: {valid_temp.mean():.2f}°C")
    
    # WM1 in December
    wm1 = f[f"Aventa/{first_block}/WM1/Value"][:].flatten()
    valid_wm1 = wm1[wm1 > -999]
    print(f"\nWM1 in December file:")
    print(f"  Min: {valid_wm1.min():.2f}, Max: {valid_wm1.max():.2f}")
```

```python
NORMAL_PATH = Path("C:/Projects/win-turbine-twin/data/aventa_normal_operation_for_system_identification/")
normal_files = list(NORMAL_PATH.glob("*.hdf5"))
print(f"Found {len(normal_files)} normal operation files")
```

```python
all_wind = []
all_power = []

for file in normal_files:
    with h5py.File(file, "r") as f:
        blocks = list(f["Aventa"].keys())
        for block in blocks:
            wind = f[f"Aventa/{block}/WM2/Value"][:].flatten()
            power = f[f"Aventa/{block}/WM3/Value"][:].flatten()
            # Filter out sentinel values
            mask = (wind > -999) & (power > -999)
            all_wind.extend(wind[mask].tolist())
            all_power.extend(power[mask].tolist())

df_normal = pd.DataFrame({
    "wind_speed": all_wind,
    "power": all_power
})

print(f"Total readings: {len(df_normal)}")
print(f"Wind range: {df_normal['wind_speed'].min():.1f} - {df_normal['wind_speed'].max():.1f} m/s")
print(f"Power range: {df_normal['power'].min():.2f} - {df_normal['power'].max():.2f} kW")
```

```python
all_wind = []
all_power = []
skipped = 0

for file in normal_files:
    with h5py.File(file, "r") as f:
        blocks = list(f["Aventa"].keys())
        for block in blocks:
            try:
                wind = f[f"Aventa/{block}/WM2/Value"][:].flatten()
                power = f[f"Aventa/{block}/WM3/Value"][:].flatten()
                mask = (wind > -999) & (power > -999)
                all_wind.extend(wind[mask].tolist())
                all_power.extend(power[mask].tolist())
            except KeyError:
                skipped += 1
                continue

df_normal = pd.DataFrame({
    "wind_speed": all_wind,
    "power": all_power
})

print(f"Total readings: {len(df_normal)}")
print(f"Skipped blocks: {skipped}")
print(f"Wind range: {df_normal['wind_speed'].min():.1f} - {df_normal['wind_speed'].max():.1f} m/s")
print(f"Power range: {df_normal['power'].min():.2f} - {df_normal['power'].max():.2f} kW")
```

```python
# Filter to operating range only - remove negatives and above cut-off
df_clean = df_normal[
    (df_normal["wind_speed"] >= 0) &
    (df_normal["wind_speed"] <= 14) &
    (df_normal["power"] >= 0)
].copy()

print(f"Clean readings: {len(df_clean)}")

plt.figure(figsize=(10, 6))
plt.scatter(df_clean["wind_speed"], df_clean["power"], 
            alpha=0.05, s=2, color="steelblue")
plt.xlabel("Wind speed (m/s)")
plt.ylabel("Power output (kW)")
plt.title("AV-7 power curve — normal operation data")
plt.axvline(x=2, color="green", linestyle="--", linewidth=1, label="Cut-in (2 m/s)")
plt.axvline(x=14, color="red", linestyle="--", linewidth=1, label="Cut-off (14 m/s)")
plt.axhline(y=7, color="orange", linestyle="--", linewidth=1, label="Rated (7 kW)")
plt.legend()
plt.tight_layout()
plt.savefig("power_curve_scatter.png", dpi=150)
plt.show()

print("Plot saved as power_curve_scatter.png")
```

```python
# Check what WM5 rotor status values exist
print("WM5 unique values in normal data:")
all_status = []
for file in normal_files:
    with h5py.File(file, "r") as f:
        blocks = list(f["Aventa"].keys())
        for block in blocks:
            try:
                status = f[f"Aventa/{block}/WM5/Value"][:].flatten()
                all_status.extend(status[status > -999].tolist())
            except KeyError:
                continue

import numpy as np
unique_status = np.unique(all_status)
print(unique_status)
```

```python
# Load wind, power and status together for one file
all_wind = []
all_power = []
all_status = []

for file in normal_files:
    with h5py.File(file, "r") as f:
        blocks = list(f["Aventa"].keys())
        for block in blocks:
            try:
                wind = f[f"Aventa/{block}/WM2/Value"][:].flatten()
                power = f[f"Aventa/{block}/WM3/Value"][:].flatten()
                status = f[f"Aventa/{block}/WM5/Value"][:].flatten()
                mask = (wind > -999) & (power > -999) & (status > -999)
                all_wind.extend(wind[mask].tolist())
                all_power.extend(power[mask].tolist())
                all_status.extend(status[mask].tolist())
            except KeyError:
                continue

df = pd.DataFrame({
    "wind_speed": all_wind,
    "power": all_power,
    "status": all_status
})

# For each status code, show mean power
print("Status code vs mean power output:")
print(df.groupby("status")["power"].agg(["mean", "count"]).round(2))
```

```python
df_running = df[df["status"] == 10].copy()

print(f"Running readings: {len(df_running)}")
print(f"Wind range: {df_running['wind_speed'].min():.1f} - {df_running['wind_speed'].max():.1f} m/s")
print(f"Power range: {df_running['power'].min():.2f} - {df_running['power'].max():.2f} kW")

plt.figure(figsize=(10, 6))
plt.scatter(df_running["wind_speed"], df_running["power"],
            alpha=0.05, s=2, color="steelblue")
plt.xlabel("Wind speed (m/s)")
plt.ylabel("Power output (kW)")
plt.title("AV-7 power curve — status 10 (generating) only")
plt.axvline(x=2, color="green", linestyle="--", linewidth=1, label="Cut-in (2 m/s)")
plt.axvline(x=14, color="red", linestyle="--", linewidth=1, label="Cut-off (14 m/s)")
plt.axhline(y=7, color="orange", linestyle="--", linewidth=1, label="Rated (7 kW)")
plt.legend()
plt.tight_layout()
plt.savefig("power_curve_running.png", dpi=150)
plt.show()
```

```python
# Filter to clean operating range
df_fit = df_running[
    (df_running["wind_speed"] >= 2.0) &
    (df_running["wind_speed"] <= 14.0) &
    (df_running["power"] >= 0)
].copy()

# Bin wind speeds in 0.5 m/s intervals, take mean power per bin
df_fit["wind_bin"] = (df_fit["wind_speed"] / 0.5).round() * 0.5
binned = df_fit.groupby("wind_bin")["power"].mean().reset_index()
binned.columns = ["wind_speed", "expected_power"]

print("Binned power curve:")
print(binned.to_string(index=False))

# Plot binned curve over scatter
plt.figure(figsize=(10, 6))
plt.scatter(df_fit["wind_speed"], df_fit["power"],
            alpha=0.02, s=2, color="steelblue", label="Raw data")
plt.plot(binned["wind_speed"], binned["expected_power"],
         color="red", linewidth=2, label="Fitted curve (binned mean)")
plt.xlabel("Wind speed (m/s)")
plt.ylabel("Power output (kW)")
plt.title("AV-7 power curve — fitted")
plt.legend()
plt.tight_layout()
plt.savefig("power_curve_fitted.png", dpi=150)
plt.show()
```
